### Задание 1:Изоморфизмы
+ Реализовать функцию (или тело функции), которая проверяет на изоморфность два слова. Пояснение: строки s и t называются изоморфными, если все вхождения каждого символа строки s можно последовательно заменить другим символом и получить строку t. Порядок символов при этом должен сохраняться, а замена — быть уникальной. Так, два разных символа строки s нельзя заменить одним и тем же символом из строки t, а вот одинаковые символы в строке s должны заменяться одним и тем же символом.   
1. Проверяем длину строк. Если разная — сразу False.
2. mapping хранит соответствие символ s → символ t.
3. used хранит символы t, которые уже заняты.
4. Проходим по строкам одновременно (zip).
5. Если символ s уже встречался — проверяем, совпадает ли его соответствие.
6. Если символ новый — убеждаемся, что символ t ещё не использован.
7. В конце возвращаем True, если конфликтов не было.

#### Оценить оптимальность решения по времени и памяти и прикрепить текст кода.    
**Временная сложность**: O(n), так как выполняется один проход по строкам длины n.Все операции со словарём и множеством выполняются за     O(1).   
**Сложность по памяти**: O(n), поскольку используется словарь и множество для хранения соответствий символов.


```python
s = 'paper' 
t = 'title' 


def is_isomorphic(s, t):
    if len(s) != len(t):
        return False

    mapping = {}
    used = set()

    for c1, c2 in zip(s, t):
        if c1 in mapping:
            if mapping[c1] != c2:
                return False
        else:
            if c2 in used:
                return False
            mapping[c1] = c2
            used.add(c2)

    return True

```

```python
# проверка функции 
is_isomorphic(s, t)
```

### Задание 2: Натуральная последовательность
+ Реализовать функцию (или тело функции), которая находит единственное отсутствующее число из последовательности натуральных чисел 1,2,…,n.   
  **Код делает следующее**:
  1. определяет n = len(nums) + 1, то есть предполагаемую длину полной последовательность
  2. вычисляет ожидаемую сумму чисел от 1 до n по формуле $\frac{n(n+1)}{2}$.
  3.  считает реальную сумму массива, разница между ними — пропущенное число.
+ Временная сложность алгоритма: O(n), так как происходит один проход по строкам. Сложность по памяти: O(n), поскольку используется словарь и множество для хранения соответствий символов.

```python
def missing_number(nums):
    n = len(nums) + 1
    expected_sum = n * (n + 1) // 2
    actual_sum = sum(nums)

    return expected_sum - actual_sum
```

```python
# проверка функции 
nums = [1, 2, 3, 4, 5, 6, 8, 9, 10, 11]
missing_number(nums)
```

### Задание 3: Факторизация
Реализовать функцию (или тело функции), которая при введении натурального числа n разбивает его на простые множители (представить его в виде простых чисел).
Оценка сложности: Время-O(n) Память-O(1)


```python
def prime_factors(n):
    factors = []
    divisor = 2

    while n > 1:
        if n % divisor == 0:
            factors.append(divisor)
            n = n // divisor
        else:
            divisor += 1
    return factors
   
```

```python
# проверка функции 
prime_factors(56)
```

```python
# проверка функции 
prime_factors(84)
```

```python
# проверка функции 
prime_factors(36)
```
